# 📓 Introducción a Blockchain
### ¿Cómo funciona la tecnología blockchain?

## 🧠 ¿Qué es Blockchain?

Blockchain es una tecnología que permite almacenar información de forma segura, descentralizada e inmutable.

En lugar de depender de un servidor central, la información se distribuye en múltiples computadoras llamadas nodos.

```python
import hashlib

def crear_hash(data):
    return hashlib.sha256(data.encode()).hexdigest()

dato = "Ana paga a Luis 10 BTC"
print("Hash generado:", crear_hash(dato))
```

## ⛓️ Estructura de un bloque

Cada bloque contiene:
- Datos
- Hash propio
- Hash del bloque anterior

Esto permite que los bloques estén conectados formando una cadena.

```python
import hashlib

class Bloque:
    def __init__(self, data, hash_anterior):
        self.data = data
        self.hash_anterior = hash_anterior
        self.hash = self.crear_hash()

    def crear_hash(self):
        contenido = self.data + self.hash_anterior
        return hashlib.sha256(contenido.encode()).hexdigest()

bloque1 = Bloque("Primera transacción", "0")
bloque2 = Bloque("Segunda transacción", bloque1.hash)

print("Hash bloque 1:", bloque1.hash)
print("Hash bloque 2:", bloque2.hash)
```

## 🔐 Seguridad en Blockchain

Si se modifica un bloque, su hash cambia automáticamente.

Esto rompe la cadena, haciendo evidente cualquier intento de alteración.

```python
bloque_prueba = Bloque("Dato original", "12345")
print("Hash original:", bloque_prueba.hash)

# Modificar datos
bloque_prueba.data = "Dato modificado"
print("Nuevo hash:", bloque_prueba.crear_hash())
```

## ⚙️ Minería (Proof of Work)

La minería consiste en resolver problemas matemáticos para validar bloques.

Se busca un hash que comience con ciertos ceros.

```python
import hashlib

nonce = 0

while True:
    texto = "transaccion" + str(nonce)
    hash_resultado = hashlib.sha256(texto.encode()).hexdigest()

    if hash_resultado.startswith("000"):
        print("Nonce encontrado:", nonce)
        print("Hash válido:", hash_resultado)
        break

    nonce += 1
```

## 🌐 Descentralización

En blockchain no existe un único servidor.

La información se distribuye en múltiples nodos, lo que la hace más segura y resistente a fallos.
